```python
#==> Cái dataset này anyf đang có số lượng class cần dự đoán ko cân bằng
```

```python
#Chia data ra để phân tích
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
#Cân bằng lại class bằng cách xây dựng 1 mẫu dataset chứa sự phân phối tương tự như giao dịch hợp pháp và giao dịch ko hợp pháp
```

```python
#Under-Sampling: Lấy mẫu ít hơn từ Class legit để training
legit_sample = legit.sample(n=492)
```

```python
#Kết nối với df
new_dataset = pd.concat([legit_sample, fraud], axis=0)
#axis=0 cho row
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns='Class', axis=1)
y = new_dataset['Class']
```

```python
print(X)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=2, test_size=0.2, stratify=y)
```

```python
model = LogisticRegression()
```

```python
#Training Logistic model với training data
model.fit(X_train, y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = metrics.accuracy_score(y_train, X_train_prediction)
```

```python
print(training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = metrics.accuracy_score(y_test, X_test_prediction)
```

```python
print(test_data_accuracy)
```
