```python
#Under-Sampling: Láº¥y máº«u Ã­t hÆ¡n tá»« Class legit Ä‘á»ƒ training
legit_sample = legit.sample(n=492)
```

```python
#Káº¿t ná»‘i vá»›i df
new_dataset = pd.concat([legit_sample, fraud], axis=0)
#axis=0 cho row
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns='Class', axis=1)
y = new_dataset['Class']
```

```python
print(X)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=2, test_size=0.2, stratify=y)
```

```python
model = LogisticRegression()
```

```python
#Training Logistic model vá»›i training data
model.fit(X_train, y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = metrics.accuracy_score(y_train, X_train_prediction)
```

```python
print(training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = metrics.accuracy_score(y_test, X_test_prediction)
```

```python
print(test_data_accuracy)
```
